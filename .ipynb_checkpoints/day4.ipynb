```python
import cv2
from time import sleep
vid = cv2.VideoCapture(0)

while True:
    flag, img = vid.read()
    if flag:
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        th, img_bw = cv2.threshold(img_gray,100,255, cv2.THRESH_BINARY)
        cv2.imshow('Preview',img_bw)
        key = cv2.waitKey(1)
        if key == ord('x'):
            break
    else:
            break
    sleep(0.1)
vid.release()
cv2.destroyAllWindows()
```

```python
import cv2
from time import sleep
vid = cv2.VideoCapture(0)

while True:
    flag, img = vid.read()
    if flag:
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        x1, y1, w, h=(200,400,200,200)
        img_cropped = img[y1:y1+h, x1:x1+w, :]

        cv2.rectangle(
             img,
             pt1=(x1,y1), pt2=(x1+w,y1+h),
             color=(0,0,225), thickness=10
             )
        
        cv2.imshow('Preview',img)
        key = cv2.waitKey(1)
        if key == ord('x'):
            break
    else:
            break
    sleep(0.1)
vid.release()
cv2.destroyAllWindows()
```

```python
#face detection
import cv2
from time import sleep

fd=cv2.CascadeClassifier(
     cv2.data.haarcascades +
     'haarcascade_frontalface_default.xml'
)

vid = cv2.VideoCapture(0)

while True:
    flag, img = vid.read()
    if flag:
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        x1, y1, w, h=(200,400,200,200)
         
        faces = fd.detectMultiScale(img_gray,1.1,5)
        for x1,y1,w,h in faces:
            img_cropped = img[y1:y1+h, x1:x1+w, :]

            cv2.rectangle(
                img,
                pt1=(x1,y1), pt2=(x1+w,y1+h),
                color=(0,0,225), thickness=10
                )
        
        cv2.imshow('Preview',img)
        key = cv2.waitKey(1)
        if key == ord('x'):
            break
    else:
            break
    sleep(0.1)
vid.release()
cv2.destroyAllWindows()
```
